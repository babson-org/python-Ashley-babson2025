## Roman Numerals to Integer

**Question**

Write a function roman_to_int(roman) that converts a valid Roman numeral string into its integer value.  
Use the standard symbols I, V, X, L, C, D, M and standard subtractive combinations (e.g., IV = 4, IX = 9).  
```

```

````python
'''
Computational Thinking

I = 1
V = 5
X = 10
L = 50
C = 100
D = 500
M = 1000

Add when smaller comes after bigger: VI = 5 + 1 = 6
Subtract when smaller comes before bigger: IV = 5 − 1 = 4
No more than three of the same in a row: III is OK, but not IIII (use IV).

Only I, X, and C subtract:
I before V or X (IV, IX)
X before L or C (XL, XC)
C before D or M (CD, CM)

Example:
MCMXCIV = 1000 + 900 + 90 + 4 = 1994

use a dictionary for the numerals 

'''
````

````python
'''
pseudo code - add structure




'''
````

````python
# actual code 3
def roman_to_int(roman:str) -> int:
    romans = {"I:1",
                "V:5",
                "X:10",
                "L:50",
                "C:100",
                "D:500",
                "M:1000"}
    #compare current symbol to next one, decide whether to add or subtract 
    total = 0
    i = 0
    length = len(roman)
    while i < length - 1:                                               #stay in bound
        if i + 1 < length:                                          #check for subtraction
            if romans[roman[i]] < romans[roman[i+1]]:
                total += romans[roman[i+1]]- romans[roman[i]]
            else:
                total += romans[roman[i]]
                i += 1                                              #skip over roamn[i+1]
        else: 
            total += romans[roman[i]]                               #get final numeral 
    return total

        
````

````python
# actual code 5

def num_col(n):
    result = ''
    while n > 0:
        n-=1
        result = chr(ord("A") + (n % 26)) + result
        n //= 26
    return result 



````

## Word Wrap at Fixed Width

**Question**

Write a function `wrap(text, width=40)` that returns a string where  
newlines (`"\\n"`) have been inserted so that no line exceeds `width`  
characters. Lines should break only at spaces between words.
```

```

````python
'''
Computational Thinking

- split the string for every word
- first word goes in string, is len > 40, if not, append next word and add space
- build string until next word goes past 40
- then add \n so you get new line
- then define line as empty and start over 

'''
````

````python
'''
pseudo code - add structure






'''
````

````python
# actual code 7

def wrap(text, width = 40):
    words = text.split()
    lines = []
    line = ''

    for w in words:
        if not line:
            line = w
        elif len(line) + 1 = len(w) <= width:
            line += ' ' + w 
        else:
            lines.append(line)
            line = w

    if line:
        lines.append(line)
    
    return'\n'.join(lines)



````
